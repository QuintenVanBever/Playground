```python
import pandas as pd
import csv
import json
```

```python
# Generate checklist dataset
f = open('files/pokedex.json')
data = json.load(f)

# Sort based on nationalDexNumber 
def sort_by_key(list):
    return list['nationalDexNumber']

sortedData = sorted(data, key=sort_by_key)

checklistdata = []
for index, i in enumerate(sortedData):

    # GET base values from json file
    mon = {
        "id": index,
        "natDexNumber": i['nationalDexNumber'],
        "name": i['name'],
    }
    # 
    if  i['formName']:
        mon['formType'] = i['formName']
        mon['formName'] = i['name'] +'-'+i['formName']
        mon['formNameAlt'] = i['formName'] + ' ' + i['name']
        mon['isBase'] = False
        mon['spriteID'] = (str(i['nationalDexNumber']) + '-' + i['formName'].lower()).replace(' ', '-')
    else: 
        mon['isBase'] = True
        mon['spriteID'] = i['nationalDexNumber']
        
    mon['boxable']= True
    checklistdata.append(mon)
print(len(checklistdata))

jsondata = json.dumps(checklistdata)

with open("files/checklist-pokedex.json", "w")as outfile:
    outfile.write(jsondata)
    
```
